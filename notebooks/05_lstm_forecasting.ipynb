# **BackTest on 3 days**

```python
# 1) IMPORTS & CONFIG

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.preprocessing import MinMaxScaler

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)

# ── Config ──
SEQ_LEN      = 192       # Input window: 2 days (192 × 15min)
BATCH_SIZE   = 64
MAX_EPOCHS   = 150       # Raised — EarlyStopping will cut short if needed
PATIENCE     = 20        # FIX: was 10, LSTMs need more room to settle
MIN_EPOCHS   = 30        # FIX: don't allow early stop before epoch 30
VAL_MONTHS   = 2         # FIX: explicit 2-month val window from end of train
LSTM_UNITS_1 = 128
LSTM_UNITS_2 = 64
DROPOUT_RATE = 0.2
HORIZONS     = {"24h": 96, "48h": 192, "72h": 288}
N_BACKTEST   = 5

print("Config loaded.")
print(f"  SEQ_LEN   = {SEQ_LEN} steps | VAL_MONTHS = {VAL_MONTHS} | PATIENCE = {PATIENCE}")


# 2) DATA LOADING
df = pd.read_parquet("nrldc_cleaned.parquet")
df.columns = ["load"]
df = df.sort_index()

print(f"\nDataset: {len(df)} rows  |  {df.index[0]}  →  {df.index[-1]}")


# 3) TRAIN / VAL / TEST SPLIT

test_cutoff = df.index[-1] - pd.DateOffset(months=3)
val_cutoff  = test_cutoff  - pd.DateOffset(months=VAL_MONTHS)

train_df = df[df.index <  val_cutoff]
val_df   = df[(df.index >= val_cutoff) & (df.index < test_cutoff)]
test_df  = df[df.index >= test_cutoff]

print(f"\nSplit summary:")
print(f"  Train : {len(train_df)} rows  ({train_df.index[0].date()} → {train_df.index[-1].date()})")
print(f"  Val   : {len(val_df)}  rows  ({val_df.index[0].date()}  → {val_df.index[-1].date()})")
print(f"  Test  : {len(test_df)} rows  ({test_df.index[0].date()}  → {test_df.index[-1].date()})")

# Warn if the known summer gap (approx May–Aug 2025) overlaps with val
val_start_month = val_df.index[0].month
val_end_month   = val_df.index[-1].month
if any(m in range(5, 9) for m in range(val_start_month, val_end_month + 1)):
    print("\n  ⚠  WARNING: Val window overlaps with peak summer months.")
    print("     If the ~67-day gap falls here, val_loss will be noisy.")
    print("     Consider increasing VAL_MONTHS to push val window earlier.")
else:
    print("\n  ✓  Val window looks gap-free.")


# 4) FEATURE ENGINEERING  (cyclical time encoding)

def add_time_features(series: pd.Series) -> pd.DataFrame:
    feat = pd.DataFrame({"load": series})
    hour  = series.index.hour + series.index.minute / 60
    feat["hour_sin"]  = np.sin(2 * np.pi * hour / 24)
    feat["hour_cos"]  = np.cos(2 * np.pi * hour / 24)
    dow = series.index.dayofweek
    feat["dow_sin"]   = np.sin(2 * np.pi * dow / 7)
    feat["dow_cos"]   = np.cos(2 * np.pi * dow / 7)
    month = series.index.month
    feat["month_sin"] = np.sin(2 * np.pi * month / 12)
    feat["month_cos"] = np.cos(2 * np.pi * month / 12)
    return feat

full_feat = add_time_features(df["load"])
N_FEATURES = full_feat.shape[1]
print(f"\nFeatures ({N_FEATURES}): {list(full_feat.columns)}")

# 5) SCALING  (fit on train only)

train_len = len(train_df)
val_len   = len(val_df)

load_scaler = MinMaxScaler(feature_range=(0, 1))
load_scaler.fit(train_df[["load"]])

full_feat_scaled = full_feat.copy()
full_feat_scaled["load"] = load_scaler.transform(full_feat[["load"]])

print(f"\nScaler fit on train only.")
print(f"  Load range in train: {load_scaler.data_min_[0]:.0f} – {load_scaler.data_max_[0]:.0f} MW")


# 6) SEQUENCE BUILDER

def build_sequences(feat_array: np.ndarray, seq_len: int, horizon: int):
    """
    Sliding window → (X, y) pairs.
    X : (n, seq_len, n_features)
    y : (n, horizon)  — load column only
    """
    X, y = [], []
    for i in range(len(feat_array) - seq_len - horizon + 1):
        X.append(feat_array[i : i + seq_len])
        y.append(feat_array[i + seq_len : i + seq_len + horizon, 0])
    return np.array(X, dtype=np.float32), np.array(y, dtype=np.float32)


# 7) CUSTOM EARLY STOPPING WITH MIN_EPOCHS GUARD

class MinEpochEarlyStopping(EarlyStopping):
    """
    EarlyStopping that refuses to fire before `min_epochs` epochs.
    Fixes the root cause of v1's 'best epoch: 3' problem.
    """
    def __init__(self, min_epochs: int, **kwargs):
        super().__init__(**kwargs)
        self.min_epochs = min_epochs

    def on_epoch_end(self, epoch, logs=None):
        if epoch < self.min_epochs:
            return          # Do nothing — too early to stop
        super().on_epoch_end(epoch, logs)


# 8) MODEL BUILDER

def build_lstm_model(seq_len: int, n_features: int, horizon: int) -> tf.keras.Model:
    model = Sequential([
        LSTM(LSTM_UNITS_1, input_shape=(seq_len, n_features), return_sequences=True),
        Dropout(DROPOUT_RATE),
        LSTM(LSTM_UNITS_2, return_sequences=False),
        Dropout(DROPOUT_RATE),
        Dense(horizon)
    ])
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
        loss="mse"
    )
    return model


# 9) TRAIN + BACKTEST (per horizon)

feat_array = full_feat_scaled.values.astype(np.float32)
series     = df["load"]

# Index positions for split boundaries
val_start_idx  = train_len
val_end_idx    = train_len + val_len
test_start_idx = val_end_idx

summary_results = []

for horizon_name, horizon_steps in HORIZONS.items():

    print(f"\n{'='*55}")
    print(f"  HORIZON: {horizon_name}  ({horizon_steps} steps = {horizon_steps*15//60}h)")
    print(f"{'='*55}")

    # ── 9a) Build sequences ──
    # Train sequences: built from train portion only
    train_array = feat_array[:val_start_idx]
    X_train, y_train = build_sequences(train_array, SEQ_LEN, horizon_steps)

    # Val sequences: built from [end-of-train - SEQ_LEN ... end-of-val]
    # We include SEQ_LEN steps of train tail so val sequences have full history
    val_array = feat_array[val_start_idx - SEQ_LEN : val_end_idx]
    X_val, y_val = build_sequences(val_array, SEQ_LEN, horizon_steps)

    print(f"  Train sequences : {X_train.shape}")
    print(f"  Val   sequences : {X_val.shape}")

    # ── 9b) Build model ──
    model = build_lstm_model(SEQ_LEN, N_FEATURES, horizon_steps)

    callbacks = [
        MinEpochEarlyStopping(
            min_epochs=MIN_EPOCHS,          # FIX: don't stop before epoch 30
            monitor="val_loss",
            patience=PATIENCE,              # FIX: was 10, now 20
            restore_best_weights=True,
            verbose=1
        ),
        ReduceLROnPlateau(
            monitor="val_loss",
            factor=0.5,
            patience=8,                     # FIX: was 5, now 8
            min_lr=1e-6,
            verbose=1
        )
    ]

    # ── 9c) Train ──
    history = model.fit(
        X_train, y_train,
        epochs=MAX_EPOCHS,
        batch_size=BATCH_SIZE,
        validation_data=(X_val, y_val),    # FIX: explicit val, not split=0.1
        callbacks=callbacks,
        verbose=0
    )

    best_epoch = np.argmin(history.history["val_loss"]) + 1
    best_val   = min(history.history["val_loss"])
    total_epochs = len(history.history["loss"])
    print(f"\n  Best epoch : {best_epoch} / {total_epochs}")
    print(f"  Best val_loss : {best_val:.6f}")

    if best_epoch < 10:
        print("  ⚠  Best epoch still very low — check val window for data gaps.")

    # ── 9d) Training curve ──
    plt.figure(figsize=(9, 3))
    plt.plot(history.history["loss"],     label="Train Loss", color="steelblue")
    plt.plot(history.history["val_loss"], label="Val Loss",   color="coral")
    plt.axvline(best_epoch - 1, color="red", linestyle="--",
                label=f"Best epoch ({best_epoch})")
    plt.title(f"LSTM Training Curve — {horizon_name}  |  Best epoch: {best_epoch}")
    plt.xlabel("Epoch")
    plt.ylabel("MSE Loss (scaled)")
    plt.legend()
    plt.tight_layout()
    plt.show()

    # ── 9e) Backtest: 5 windows ──
    results = []

    for i in range(N_BACKTEST):

        idx = test_start_idx + i * horizon_steps

        x_input = feat_array[idx - SEQ_LEN : idx]
        x_input = x_input.reshape(1, SEQ_LEN, N_FEATURES)

        preds_scaled = model.predict(x_input, verbose=0)[0]
        preds_mw     = load_scaler.inverse_transform(
            preds_scaled.reshape(-1, 1)
        ).flatten()

        actual_mw = series.iloc[idx : idx + horizon_steps].values
        min_len   = min(len(preds_mw), len(actual_mw))
        preds_mw  = preds_mw[:min_len]
        actual_mw = actual_mw[:min_len]

        mae  = mean_absolute_error(actual_mw, preds_mw)
        rmse = np.sqrt(mean_squared_error(actual_mw, preds_mw))
        mape = np.mean(np.abs((actual_mw - preds_mw) / actual_mw)) * 100

        results.append([i + 1, mae, rmse, mape])

        plt.figure(figsize=(10, 3))
        plt.plot(actual_mw, label="Actual",    color="steelblue")
        plt.plot(preds_mw,  label="Predicted", color="coral", linestyle="--")
        plt.title(f"LSTM {horizon_name} | Window {i+1} | MAPE: {mape:.2f}%")
        plt.xlabel("Steps (15-min intervals)")
        plt.ylabel("Load (MW)")
        plt.legend()
        plt.tight_layout()
        plt.show()

    # ── 9f) Per-horizon summary ──
    res = pd.DataFrame(results, columns=["Window", "MAE", "RMSE", "MAPE (%)"])
    print(f"\nBacktest Summary — {horizon_name}")
    print(res.to_string(index=False))

    avg_mape = res["MAPE (%)"].mean()
    print(f"\n  Avg MAPE : {avg_mape:.2f}%")

    summary_results.append([
        horizon_name,
        res["MAE"].mean(),
        res["RMSE"].mean(),
        avg_mape
    ])


# 10) FINAL COMPARISON TABLE

final = pd.DataFrame(
    summary_results,
    columns=["Horizon", "Avg MAE", "Avg RMSE", "Avg MAPE (%)"]
)

print("\n" + "="*55)
print("  LSTM v2 — FORECAST HORIZON COMPARISON")
print("="*55)
print(final.to_string(index=False))

xgb_results = {
    "24h": 1.69,
    "48h": 2.11,
    "72h": 2.74,
}

print("\n" + "="*60)
print("  HEAD-TO-HEAD: LSTM v2 vs XGBoost (Avg MAPE %)")
print("="*60)
print(f"{'Horizon':<10} {'XGBoost':>12} {'LSTM v2':>12} {'Delta':>10} {'Winner':>12}")
print("-"*60)

for _, row in final.iterrows():
    h     = row["Horizon"]
    xmape = xgb_results[h]
    lmape = row["Avg MAPE (%)"]
    delta = lmape - xmape
    sign  = "+" if delta > 0 else ""
    winner = "LSTM ✓" if lmape < xmape else "XGBoost ✓"
    print(f"{h:<10} {xmape:>11.2f}% {lmape:>11.2f}% {sign+f'{delta:.2f}%':>10} {winner:>12}")

print("\nDone.")
```

# **Executive Summary — Load Forecasting Model Evaluation**

## Dataset & Setup
- Data Duration: April 2024 to March 2026 (~59,000 rows, 15-minute intervals)
- Features Used (7): Load and time-based cyclic features (hour, day of week, month)
- Train/Validation/Test Split: Clean and gap-free
- Model: LSTM v2 with sequence length of 192 (past 48 hours)

---

## Model Performance (LSTM v2)

### Forecast Accuracy by Horizon
| Horizon | Avg MAPE | Observation |
|--------|----------|------------|
| 24h    | 2.30%    | Strong and stable |
| 48h    | 2.31%    | Slight degradation |
| 72h    | 2.99%    | Noticeable drop |

Key Insight:
- Performance degrades as forecast horizon increases
- Error remains under 3%, which is acceptable for load forecasting tasks

---

## Training Behavior Insights
- Early stopping triggered at very early epochs (best epoch between 6 and 9)
- Learning rate reduced multiple times, indicating quick convergence
- Warning flagged: "Best epoch very low — check validation window"

Interpretation:
- Model is learning patterns very quickly
- Indicates either strong seasonality in data or limited model complexity (possible underfitting)

---

## Head-to-Head Comparison (LSTM v2 vs XGBoost)

| Horizon | XGBoost | LSTM v2 | Delta | Winner |
|--------|--------|---------|-------|--------|
| 24h    | 1.69%  | 2.30%   | +0.61% | XGBoost |
| 48h    | 2.11%  | 2.31%   | +0.20% | XGBoost |
| 72h    | 2.74%  | 2.99%   | +0.25% | XGBoost |

Conclusion:
- XGBoost consistently outperforms LSTM across all forecast horizons

---

## Critical Observations

### 1. XGBoost Superiority
- Better accuracy and stability across all horizons
- Well-suited for structured/tabular data with engineered features

### 2. Limited Utilization of LSTM
- Only basic time-based features used
- Missing important external variables such as:
  - Weather data
  - Holidays
  - Demand spikes or special events

This limits the ability of LSTM to leverage temporal dependencies effectively

### 3. Weakness in Long Horizon (72h)
- Significant increase in error
- Instability observed in later backtest windows
- Indicates difficulty in long-term forecasting

---

## Final Verdict

### Recommended Model for Deployment: XGBoost
- Higher accuracy
- More stable performance
- Easier to deploy and maintain

### Status of LSTM
- Performs reasonably well but not optimal
- Suitable for experimentation or enhancement with additional features
